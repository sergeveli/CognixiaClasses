Let's walk through each table and explain.

**Song Table**

First we take the name, track_number, disc number, duration_ms, time_signature, year, and realease_date as the first logical grouping. 

Next we include a song_id column to uniquely identify each entity. 

Finally we include the album_id to connect albums with their numerous songs.

**Attributes Table**

Here we have the remaining relevant attributes including explicit, danceability, energy, loudness, mode, speechiness, acousticness, instrumentalness, liveness, valence, tempo. **Album Artist Table**

This table connects each artist to an album and all other tables transitively. The artist table can logically be connected to either the song or album tables, but during this exercise we will only connect the album and artist table

# **Data Cleaning**

*Please run the cells below to read the csv into a dataframe.*





```python
import pandas as pd
```

```python
path = '/Users/veli/Desktop/DataEngineering/W3D3/SpotifyDataPandas.csv'
```

```python
spotify_df = pd.read_csv(path)
```

```python
spotify_df.head()
```

*Lets start by getting rid of some of the unneccessary columns (columns:Unnamed: 0, id, album_id,artist_ids,key) and renaming the index to song_id*

```python
clean_spotify_df = spotify_df.drop(columns=['Unnamed: 0', 'id', 'album_id', 'artist_ids', 'key'])

clean_spotify_df.index.name = 'song_id'

print(clean_spotify_df.head())
```

image.png

*Great! Removing irrelevant data as soon as possible can make the cleaning process easier to manage. Let's begin breaking the table down into the smaller dataframes decided with our schema.*

---

**Album Table**
Here we need to list each unique album in a dataframe.

```python
album_table = spotify_df[['album']].drop_duplicates()
album_table.index.name = 'album_id'

print(album_table)
```

image.png


---


**Song Table**

includes : name, track_number, disc_number, duration_ms, time_signature, year, release_date, and album_id

*Make sure there is a song_id column to identify each song as well*

*Note: The album_id is from the album table*

```python
# Select the columns for the song_table DataFrame
song_cols = ['name', 'track_number', 'disc_number', 'duration_ms',
             'time_signature', 'year', 'release_date']

# Create a new DataFrame with the selected columns
song_table = clean_spotify_df[song_cols]

# Create a new column 'album_id' in the song_table DataFrame using the 'album' column
song_table['album_id'] = clean_spotify_df['album'].map(
    album_table.reset_index().set_index('album')['album_id'])

# Display the song_table DataFrame
print(song_table)
```

image.png



---

**Attributes Table**

includes: attribute_id, song_id, explicit, danceability, energy, loudness, mode, speechiness, acousticness, instrumentalness, liveness, valence, tempo

```python
# Reset the index to ensure that 'song_id' is a column and not an index
clean_spotify_df = clean_spotify_df.reset_index()

# Select the columns for the attributes table DataFrame
attr_cols = ['explicit', 'danceability', 'energy', 'loudness', 'mode', 'speechiness',
             'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'song_id']

# Create a new DataFrame with the selected columns
attr_table = clean_spotify_df[attr_cols]

# Add a new 'attribute_id' column
attr_table.insert(0, 'attribute_id', range(1, len(attr_table) + 1))

# Set the 'attribute_id' column as the index
attr_table.set_index('attribute_id', inplace=True)

# Display the attr_table DataFrame
print(attr_table)
```
